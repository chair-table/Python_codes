```python
import requests
from bs4 import BeautifulSoup
import time

url = "https://www.worldometers.info/coronavirus/"

time.sleep(1)
r = requests.get(url)
doc = BeautifulSoup(r.text, "html.parser")
```

# Calculation

```python
### Bangladesh situation
def calculation(country_you_want_to_search):
    data = []
    table = doc.findAll('table', attrs={'id':'main_table_countries_today'})
    for row in table:
        cells = row.findAll('td' )

        for cell in cells:
            data.append(cell)

    dd=[]
    for x,y in enumerate(data):
        dd.append((list(y.text)))


    List1 = []
    for i in country_you_want_to_search:
        List1.append(i)


    if List1 in dd:
        index = dd.index(List1)
    else:
        index = -1
    #print(index)


    country = list(data[index])[0]
    total_case = int(list(data[index+1])[0].replace(',',''))


    try:
        new_case = int(list(data[index+2])[0].replace(',',''))
    except:
        new_case = 0 

    try:
        total_death = int(list(data[index+3])[0].replace(',',''))
    except:
        total_death = 0 

    try:
        new_death = int(list(data[index+4])[0].replace(',',''))
    except:
        new_death = 0 

    try:
        total_recovered = int(list(data[index+5])[0].replace(',',''))
    except:
        total_recovered = 0 

    active_case = int(list(data[index+6])[0].replace(',',''))

    try:
         critical_case = int(list(data[index+7])[0].replace(',',''))
    except:
         critical_case = 0



    death_rate = ((total_death)/(total_case))*100
    recovery_rate = ((total_recovered)/(total_case))*100
    spreading_rate = ((new_case)/(active_case))*100

    #print("Country: " + country_you_want_to_search )       
    df = {"total_case" : total_case,
    "total_death" : total_death,
    "death_rate(%)" : death_rate,
    "recovery_rate(%)" : recovery_rate,
    "total_recovered" : total_recovered,
    "active_case" : active_case,
    "critical_case": critical_case,
    "new_case" : new_case,
    "new_death" : new_death}
    
    
    
    def printing():
        print("\n\nCountry: " + country_you_want_to_search + "         (" + "Serial (by the total affected number): " + str((index/10)+1) + ")\n" )   
        print("\nSpreading Rate: " + str(spreading_rate)+ "\n")
        for i in df:
            if i == "recovery_rate(%)" or i == "death_rate(%)":
                print("\n")
            print(i + ": " +str(df[i]))
            if i == "recovery_rate(%)" or i == "death_rate(%)":
                print("\n")
    
    return printing()

 
```
